```python
# Plotting code

import numpy as np
import matplotlib.pyplot as plt

points = np.array([(2, 3), (0, -1), (-1, -6)])

fig = plt.figure(figsize=(10, 8))
xs = np.linspace(-3, 4, 100)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Composition of individual Lagrange polynomials")
plt.plot(xs, np.zeros_like(xs), label="$x$-axis")
plt.plot(xs, fl0(xs), label="$f(x_0)L_0(x)$")
plt.plot(xs, fl1(xs), label="$f(x_1)L_1(x)$")
plt.plot(xs, fl2(xs), label="$f(x_2)L_2(x)$")
plt.plot(xs, P(xs), label="$P(x)$")
plt.plot(points[:, 0], points[:, 1], 'bo', label="Sample points")
plt.legend()
plt.show()
```
