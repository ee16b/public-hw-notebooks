Recall that for any transfer function $H(\omega)$, the function takes in angular frequencies $\omega \in \mathbb{R}$ and will output a complex number $H(\omega) \in \mathbb{C}$. For this reason, we need two plots to describe $H(\omega)$. The first plot is the magnitude $|H(\omega)|$. The second plot is the phase $\angle H(\omega)$.

In order to implement this, we have one general function `def gen_tf` that will make the plots. See the documentation and example below to use it. 

```python
import numpy as np
import matplotlib.pyplot as plt

# This function will generate the magnitude (and optionally the phase) 
# of the inputed transfer function. 
# Inputs: 
#     H (function) - The transfer function of interest
#     w_range (length 2 tuple) - The function will plot w in [10 ** w_range[0], 10 ** w_range[1]]
#     phase (boolean) - Default is false, optionally plots the phase response
# Outputs:
#     None, just plots the function.
def gen_tf(H, w_range, phase=False):
    w = np.logspace(w_range[0], w_range[1], num=1000)
    
    plt.grid()
    plt.title('Magnitude Plot: $|H(\omega)|$')
    plt.xlabel('$\omega$')
    plt.ylabel('$|H(\omega)|$')
    plt.loglog(w, np.abs(H(w)))
    
    if phase:
        plt.figure()
        plt.grid()
        plt.title('Phase Plot: Phase of $H(\omega)$')
        plt.xlabel('$\omega$')
        plt.ylabel('Phase of $H(\omega)$')
        plt.semilogx(w, np.angle(H(w)) * 180 / np.pi)
```

```python
# EXAMPLE: Let's plot w in the range 10 ** -2 to 10 ** 5:

#Define your transfer function
def H_example(w):
    return 1 / (1 + 1j * w)

#Pass in your transfer function and the log range to gen_tf
gen_tf(H_example, (-2, 5), phase=True)
```
